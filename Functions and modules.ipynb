### 3. Passing the functions as a argument

```python
# creating fucntion which we are passing to the another fucntion as a argument.
def add(a,b):
    return a+b
# above is the first function as argument for second function
def square(c):
    return c*c
```

call the square function

```python
square(add(2,3))
```

In above **sqaure** function we passed the **add** function as argument.In **add** function we passed the two argument which 2,3. We get the final result as a 25.
